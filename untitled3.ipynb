```python
"""Trecho de código responsável pelo cadastro de um novo produto"""
def cadastrar_produto(estoque):
  #Obtém as informações do produto atual (Leitura)
  codigo = int(input("Digite o codigo do produto: "))
  categoria = input("Digite a categoria do produto: ")
  nome = input("Digite o nome do produto: ")
  descricao = input("Digite uma descrição para o produto: ")
  preco = float(input("Digite o preco do produto: "))

  #Adiciona um produto no estoque (Cadastro)
  produto = {"codigo": codigo, "produto": nome, "info": descricao, "preco": preco}
  if categoria not in estoque:
    estoque[categoria] = []
    estoque[categoria].append(produto)
  else:
    estoque[categoria].append(produto)

  #Retorna o estoque atualizado
  return estoque

#Mostrando informações do estoque
def mostrar_produto(produto):
  print("------------------------------------------")
  print("--------- Informações do produto ---------")
  print("------------------------------------------")
  print(f"Código: {produto['codigo']}")
  print(f"Produto: {produto['produto']}")
  print(f"Descrição: {produto['info']}")
  print(f"Preço: {produto['preco']}")
  print("")

#Apresenta TODOS os produtos cadastrados
def mostra_produtos(estoque):
  for categoria in estoque:
    for produto in estoque[categoria]:
      mostrar_produto(produto)

#Apresenta os produtos cadastrados com preço inferior ao limite estabelecido
def mostra_produtos_preco(estoque, preco):
  for categoria in estoque:
    for produto in estoque[categoria]:
      if produto["preco"] <= preco:
        mostrar_produto(produto)

#Produtos Star Wars
estoque = {"Eletronicos": [{
                              'codigo': 2578,
                              'produto': "Luminária LED 3D star wars, Darth Vader, stormtrooper (R2D2)",
                              'info': 'Muda de cor, 7 Cores diferentes, Pode ser utilizado com cabo USB ou pilhas, Itens Inclusos: 1 Abajur 3d, 1 -Cabo Usb, 1 Base',
                              'preco': 64.90},
                           {
                              'codigo': 2749,
                              'produto': 'Camiseta Darth Vader Star Wars Camisa Geek Filme Série Blusa',
                              'info': ' Camiseta preta Darth Vader, Instruções de cuidados: Lavagem à máquina secar na sombra',
                              'preco': 54.90},
                           {
                              'codigo': 1478,
                              'produto': 'GARRAFA MAX BABY YODA THE CHILD - STAR WARS',
                              'info': 'Altura: 29cm| Largura: 9cm| Comprimento: 9cm| Material: Plástico| BPA: Livre| Capacidade: 1,650L',
                              'preco': 84.90},
                           {
                              'codigo': 3658,
                              'produto': 'Almofada Chewbacca - Star Wars',
                              'info': 'Acompanha Enchimento e capa, Material de preenchimento: Fibra de poliéster siliconizada Marca: L3 Store',
                              'preco': 49.90}],
           "Livros":      [{
                              'codigo': 6987,
                              'produto': 'Star Wars : Um novo amanhecer',
                              'info': 'Livro Star Wars, capa comum: 1 Usado a partir de R$ 75,00, 3 Colecionador a partir de R$ 119,99',
                              'preco': 119.99},
                           ]
          }

while True:
  print("Seja bem-vindo à Star Wars Store by Última, sua loja virtual!")
  print("O que você deseja fazer: 1) Cadastrar produto, 2) Mostrar produtos, 3) Filtrar produtos por preço, 0) Sair do programa.")
  opcao = int(input("Digite a opção desejada: "))

  #Cadastrar um produto
  if opcao == 1:
    estoque = cadastrar_produto(estoque)

  #Mostrar produtos
  elif opcao == 2:
    mostra_produtos(estoque)

  #Mostrar produtos dado um limite de preço
  elif opcao == 3:
    preco_desejado = float(input("Digite o limite máximo de preço a ser buscado: "))
    mostra_produtos_preco(estoque, preco_desejado)

  #Finalizar programa
  elif opcao == 0:
    print("Agradecemos a prefêrencia!\nQue a Força esteja com você. Volte sempre!\n")
    break

  #Verifica se a opção passada é inválida
  else:
    print("Opcao inválida!")
```
